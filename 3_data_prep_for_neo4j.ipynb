```python
id_in_disease = set()
for line in open("temp/MSH_disease.json"):
    data = json.loads(line)
    id_in_disease.add(data["MSH_id"])
```

```python
diseases = {}


disease_disease = set()



#disease_file = "temp/test.txt"
disease_file = "temp/subcategories_response.txt"
for line in open(disease_file):
    items = json.loads(line)
    
    for i in items:
        son_id = i[0]
        son_name = i[1]
        parent_id = i[2]
        parent_name = i[3]

        diseases[son_id] = son_name
        disease_disease.add((son_id, parent_id))

        diseases[parent_id] = parent_name
```

```python
disease_disease
```

```python
with open("to_neo4j/disease_disease.tsv", "w") as f:
    f.write("from\tto\n")
    for i in disease_disease:
        f.write(f"{i[0]}\t{i[1]}\n")

```

```python
with open("to_neo4j/disease.tsv", "w") as f:
    f.write("MSH_id\tname\n")
    for i in diseases:
        f.write(f"{i}\t{diseases[i]}\n")
```

```python
disease_file = "temp/may_be_treated_response.tsv"

disease_drugs = set()

for line in open(disease_file):
    fields = line.strip().split("\t")
    disease_id = fields[0]
    items = json.loads(fields[1])

    for i in items:
        disease_drugs.add((disease_id, i["id"]))

with open("to_neo4j/disease_drug.tsv", "w") as f:
    f.write("from\tto\n")
    for i in disease_drugs:
        f.write(f"{i[0]}\t{i[1]}\n")
```

```python
drug_file = "temp/rxnorm_response.txt"

drugs = {}

content = "RXNORM_id\tname\tmoa\tparents\tpe\n"

for line in open(drug_file):
    items = json.loads(line.strip())

    id = items["drug_id"]
    name = items["name"]

    moa = items["moa"]
    parents = items["parents"]
    pe = items["pe"]

    content += f"{id}\t{name}\t{'|'.join([i['name'] for i in moa])}\t{'|'.join([i['name'] for i in parents])}\t{'|'.join([i['name'] for i in pe])}\n"

with open("to_neo4j/drug.tsv", "w") as f:
    f.write(content)
```
